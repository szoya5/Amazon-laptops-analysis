```python
import requests
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import re


Brand = []
OS = []
Storage_capacity = []
Discount_price = []
original_price = []
Discount = []
Rating = []
Num_of_rating = []

for i in range(1, 20):
    url = f"https://www.amazon.in/s?k=laptop&page={i}&crid=2D22M8HMQH2SZ&qid=1727958926&sprefix=laptop%2Caps%2C230&ref=sr_pg_{i}"
    print(url)
    
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.64 Safari/537.36',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        'Accept-Language': 'en-US,en;q=0.9',
        'Accept-Encoding': 'gzip, deflate, br',
        'Referer': 'https://www.justdial.com/',
        'Connection': 'keep-alive'
    }
    
    response = requests.get(url, headers=headers)
    print(response)

    soup = BeautifulSoup(response.text, 'html.parser')

    
    for product in soup.find_all("span", class_="a-size-medium a-color-base a-text-normal"):
        # Process Brand
        #brand = re.findall("^\w+", product.text)
        #Brand.append(brand[0] if brand else np.nan)
        brand = product.text.split() 
        Brand.append(brand[0] if brand[0]!="(Refurbished)" else brand[1])
       
        os_info = re.findall("(Wind?o?w?s?\s?\d{1,2}|MacBook)", product.text)
        OS.append(os_info[0] if os_info else np.nan)

        storage = re.findall("(\d+\s*GB|\d+\s*TB)\s*SSD", product.text)
        Storage_capacity.append(storage[0] if storage else np.nan)

    for product in soup.find_all("div", class_="a-row a-size-base a-color-base"):
        # Process Discount Price
        discount_price = re.findall("(₹\d{1,2},\d{3})₹.+?M.R.P", product.text)
        Discount_price.append(discount_price[0] if discount_price else np.nan)

        original = re.findall("M.R.P:\s₹(\d{1,2},\d{3})", product.text)
        original_price.append(original[0] if original else np.nan)

        discount = re.findall("\((\d{1,2})%\s", product.text)
        Discount.append(discount[0] if discount else np.nan)

    for product in soup.find_all("div", class_="a-row a-size-small"):
        
        rating = re.findall("(\d+\.\d+)", product.text)
        Rating.append(rating[0] if rating else np.nan)

        
        num_rating = re.findall("\d+\s+stars\s(\d+)", product.text)
        Num_of_rating.append(num_rating[0] if num_rating else np.nan)


max_length = max(len(Brand), len(OS), len(Storage_capacity), len(Discount_price),
                 len(original_price), len(Discount), len(Rating), len(Num_of_rating))

def pad_list(lst, max_length):
    return lst + [np.nan] * (max_length - len(lst))

Brand = pad_list(Brand, max_length)
OS = pad_list(OS, max_length)
Storage_capacity = pad_list(Storage_capacity, max_length)
Discount_price = pad_list(Discount_price, max_length)
original_price = pad_list(original_price, max_length)
Discount = pad_list(Discount, max_length)
Rating = pad_list(Rating, max_length)
Num_of_rating = pad_list(Num_of_rating, max_length)

data = {
    'Brand': Brand,
    'OS': OS,
    'Storage_capacity': Storage_capacity,
    'Discount_price': Discount_price,
    'original_price': original_price,
    'Discount': Discount,
    'Rating': Rating,
    'Num_of_rating': Num_of_rating
}

df = pd.DataFrame(data)
print(df.head())
```
